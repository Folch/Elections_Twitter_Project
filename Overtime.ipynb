```python
days=[16820,19684,21206,21981,18460,22788,18306,23475,19545,16196,18458,20374,18208, 26181,23625,19923,15854,17943,15002,27843,30807,18379,17585,15881,18802,19109,19599,32851,21707,28423,23420,8872]
```

```python
postr=32*[0]
poshil=32*[0]
negtr=32*[0]
neghil=32*[0]
posb=32*[0]
negb=32*[0]
posn=32*[0]
negn=32*[0]
```

```python
for day in range(len(days)):
    for i in range(days[day]):
        x=data.loc[i]
        if x['sentiment']=='Positive':
            if x['talk']=='trump':
                postr[day]=postr[day]+1
            elif x['talk']=='hillary':
                poshil[day]=poshil[day]+1
            elif x['talk']=='both':
                posb[day]=posb[day]+1
            else:
                posn[day]=posn[day]+1
        else:
            if x['talk']=='trump':
                negtr[day]=negtr[day]+1
            elif x['talk']=='hillary':
                neghil[day]=neghil[day]+1
            elif x['talk']=='both':
                negb[day]=negb[day]+1
            else:
                negn[day]=negn[day]+1
```

```python
plt.scatter(list(range(1,33)),postr)
plt.title('Positive for Trump')
```

```python
incomeDF = pd.read_csv("incomeUSA.csv", sep=",")
incomeDF.head()
```

```python
usCodesDF = pd.read_csv("us-states-codes-to-names.csv", sep=",")
usCodesDF.head()
```

```python
mergedDF = pd.merge(incomeDF, usCodesDF,  on = 'State')
mergedDF = mergedDF[['income', 'Abbreviation']]
mergedDF
```

```python
mapIncomeAbbreviation = mergedDF.to_dict('records')
mapIncomeAbbreviation
```

```python
loc = []
values = []
for v in mapIncomeAbbreviation:
    loc.append(v['Abbreviation'])
    values.append(v['income'])
```

```python
import plotly.plotly as py
from plotly.offline import iplot,init_notebook_mode
init_notebook_mode(connected=True)

scl = [[0.0, 'rgb(204,61,61)'],[0.5, 'rgb(255,255,255)'],\
            [1.0, 'rgb(26,128,196)']]

# df['text'] = df['state'] + '<br>' +\
#     'Beef '+df['beef']+' Dairy '+df['dairy']+'<br>'+\
#     'Fruits '+df['total fruits']+' Veggies ' + df['total veggies']+'<br>'+\
#     'Wheat '+df['wheat']+' Corn '+df['corn']

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = loc,
        z = values,
        locationmode = 'USA-states',
        text = 'Result',
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Income")
        ) ]

layout = dict(
        title = 'Income in 2016',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
iplot( fig, filename='income2016' )
```
